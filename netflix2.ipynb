GRAPH SHOWING RELATIONSHIP BETWEEN THE TWO TYPES OF FILMS (MOVIE AND TV SHOW)

```python
sns.countplot(x=data['type'])
```

```python
x=data.groupby(['type'])['type'].count()
y=len(data)
r=((x/y)).round(2)

mf_ratio = pd.DataFrame(r).T
fig, ax = plt.subplots(1,1,figsize=(6.5, 2.5))


ax.barh(mf_ratio.index, mf_ratio['Movie'], 
        color='#b20710', alpha=0.9, label='Male')
ax.barh(mf_ratio.index, mf_ratio['TV Show'], left=mf_ratio['Movie'], 
        color='#221f1f', alpha=0.9, label='Female')

ax.set_xlim(0, 1)
ax.set_xticks([])
ax.set_yticks([])
#ax.set_yticklabels(mf_ratio.index, fontfamily='serif', fontsize=11)


# movie percentage
for i in mf_ratio.index:
    ax.annotate(f"{int(mf_ratio['Movie'][i]*100)}%", 
                   xy=(mf_ratio['Movie'][i]/2, i),
                   va = 'center', ha='center',fontsize=40, fontweight='light', fontfamily='serif',
                   color='white')
for i in mf_ratio.index:
    ax.annotate(f"{int(mf_ratio['TV Show'][i]*100)}%", 
                   xy=(mf_ratio['Movie'][i]+mf_ratio['TV Show'][i]/2, i),
                   va = 'center', ha='center',fontsize=40, fontweight='light', fontfamily='serif',
                   color='white')
    ax.annotate("TV Show", 
                   xy=(mf_ratio['Movie'][i]+mf_ratio['TV Show'][i]/2, -0.25),
                   va = 'center', ha='center',fontsize=15, fontweight='light', fontfamily='serif',
                   color='white')



    






# Title & Subtitle
fig.text(0.125,1.03,'Movie & TV Show distribution', fontfamily='serif',fontsize=15, fontweight='bold')
fig.text(0.125,0.92,'We see vastly more movies than TV shows on Netflix.',fontfamily='serif',fontsize=12)  


```

GRAPH SHOWING THE DURATION OF MOVIES IN MINUTES COUNT AND TV SHOWS IN SEASON COUNT

```python
data_9 = data.query("type == 'TV Show'")
data_9 = data_9[[ "title", "duration"]]
data_9 = data_9.groupby(['duration'])["title"].count().reset_index().sort_values('title', ascending = False)
#data_9 = data_9['duration'].replace("seasons", "")
data_9 = data_9.rename(columns = {"title": "TV Shows", "duration" : "Seasons"})


data_10 = data.query("type == 'Movie'")
data_10['duration'] = data_10['duration'].fillna("0")
data_10['duration'] = data_10['duration'].str.split(" ").str[0].astype(int)





fig_show = px.histogram(data_9, x='Seasons', y='TV Shows', color_discrete_sequence=['orange'],
       title='TV Shows seasons ')
fig_Movie = px.histogram(data_10, x="duration" , nbins = 20, color_discrete_sequence=px.colors.sequential.RdBu
                  , title = "Movie Duration")


fig_Movie.show()
fig_show.show()
```
